### Store Part I results into DataFrame

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
#Configure gmaps
gmaps.configure(api_key=gkey)

#Get coordinates for the map
coordinates = df[['Lat','Lng']]

#Get humidity values for the map
humidity = df['Humidity']

#Configure layout for the map
figure_layout = {
    'width':'1000px',
    'height':'600px',
    'border':'1px solid black',
    'padding':'1px',
    'margin':'0 auto 0 auto'
}

#Plot map with heatmap layer
f = gmaps.figure(layout=figure_layout)
f.add_layer(gmaps.heatmap_layer(coordinates,humidity, point_radius=10, max_intensity = 100))
f
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
#My conditions are: Max_Temp in the range of 15 and 25 C, Wind speed <10m/s, Cloudiness <50%, Humidity <40%
hotel_df = df.loc[(df['Max_Temp']>=15) & (df['Max_Temp']<=25) & \
       (df['Wind_Speed']<=10) & (df['Cloudiness']<=50) & (df['Humidity']<=40)]
hotel_df.reset_index(inplace=True,drop=True)
hotel_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df['Hotel_Name'] = ''

#Get latitude and longitud for each city
base_url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json'
radius = 5000
business_type = 'lodging'

params = {'key':gkey,'radius':radius,'type':business_type}

for i, row in hotel_df.iterrows():
  params['location'] = str(hotel_df.loc[i,'Lat'])+','+str(hotel_df.loc[i,'Lng'])
  response = requests.get(base_url,params=params).json()

  try:
    hotel_df.loc[i,'Hotel_Name']= response['results'][0]['name']    
  except:
    hotel_df.loc[i,'Hotel_Name']= np.NaN

hotel_df
```

```python
#Remove locations where a hotel was not found
clean_hotel_df = hotel_df.dropna()
```

```python
clean_hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel_Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in clean_hotel_df.iterrows()]
locations = clean_hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
f.add_layer(gmaps.marker_layer(locations,info_box_content=hotel_info))

# Display figure
f
```
